```python
import math

# Example numbers
num1 = 4.3
num2 = 9.7
num3 = -2.8
num4 = 5.0

# Applying ceil function
print("Ceil of", num1, "is:", math.ceil(num1))  # Output: 5
print("Ceil of", num2, "is:", math.ceil(num2))  # Output: 10
print("Ceil of", num3, "is:", math.ceil(num3))  # Output: -2
print("Ceil of", num4, "is:", math.ceil(num4))  # Output: 5 (unchanged)
```
